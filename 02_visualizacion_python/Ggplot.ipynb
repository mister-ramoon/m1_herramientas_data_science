## Ggplot

```python
import pandas as pd

df = pd.read_table('iris.data', sep=',', header = None)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'type']

# Primeras observaciones
df.head()
```

### Ggplot Histograma

```python
from plotnine import *

(
    ggplot(df) +
    geom_histogram(aes(x='petal_width'))

) 
```

### Ggplot, Nube de Puntos

```python
(
     ggplot(df) +
     geom_point(aes(x='petal_width', y = 'sepal_width'))

)
```

### Agregar color a un histograma

```python
(
     ggplot(df) +
     geom_histogram(aes(x='petal_width'), color = 'blue')
)
```

```python
(
     ggplot(df) +
      geom_point(aes(x='petal_width', y = 'sepal_width'), color = 'red')

)
```

```python
(
    ggplot(df) +
     geom_histogram(aes(x='petal_width'), 
                    color = 'blue', 
                    fill = 'lightgreen', 
                    alpha = 0.2, 
                    linetype='solid')
)
```

```python
(
    ggplot(df) +
     geom_histogram(aes(x='petal_width'), 
                    color = 'blue', 
                    fill = 'lightgreen', 
                    alpha = 0.8, 
                    linetype='dashed')
)
```

### Agregar elementos textuales

```python
(
    ggplot(df) +
     geom_point(aes(x='petal_width', y = 'sepal_width'), 
                color = 'tomato',
               alpha=1,
               shape='^',
               size = 5,
               stroke = 0.2) +
    labs(title='Anchura de ptalo y spalo', 
         x = 'Anchura Ptalo',
         y = 'Anchura Spalo')
)
```

### Type

```python
df['type'].unique()
```

```python
(

     ggplot(df) +
      geom_point(aes(x='petal_width', y = 'sepal_width', color = 'type'),
                alpha=1,
                shape='.',
                size = 5,
                stroke = 0.2) +
     labs(title='Anchura de ptalo y spalo', 
          x = 'Anchura Ptalo',
          y = 'Anchura Spalo',
          color = 'Tipo')
)
```

```python
(ggplot(df)          
 + aes(x='petal_length')   
 + geom_bar() 
)
```

```python
(ggplot(df)         
 + aes(x='petal_length', fill = 'type')   
 + geom_bar(color = 'black', alpha = 0.6, show_legend=False) 
 + coord_flip() 
)
```

###  Estadsticas en una grfica

```python
(ggplot(df)         
 + aes(x='petal_length', fill = 'type')   
 + geom_bar(color = 'black', alpha = 0.6, show_legend=True) 
 + geom_vline(aes(xintercept = df.petal_length.mean()), color = 'red',
              linetype = 'dashed')
  + geom_vline(aes(xintercept = df.petal_length.median()), color = 'blue',
              linetype = 'solid')
)
```

### Facet Wrap

```python
(ggplot(df)         
 + aes(x='petal_length', fill = 'type')   
 + geom_bar(color = 'black', alpha = 0.6, show_legend=True) 
  + geom_vline(aes(xintercept = df.petal_length.mean()), color = 'red',
              linetype = 'dashed')
  + geom_vline(aes(xintercept = df.petal_length.median()), color = 'blue',
              linetype = 'solid')
 + facet_wrap('~type', ncol = 2)
 + labs(
     x = 'Longitud de ptalo',
     y = 'Valores',
     title = 'Diagrama de barras por tipo'

 )
)
```

### Incluir una recta de ajustes

(
     ggplot(df) +
      geom_point(aes(x='petal_width', y = 'petal_length', fill = 'petal_width'),
                alpha=0.5,
                shape='o',
                size = 2,
                stroke = 0.2) 
    + geom_smooth(
         aes(x='petal_width', y = 'petal_length')
    )
    + labs(title='Anchura y altura del ptalo', 
          x = 'Anchura Ptalo',
          y = 'Altura Spalo')
)

### Generar una grfica con base en sus categoras.

```python
(
    ggplot(df) +
      geom_point(aes(x='petal_width', y = 'petal_length', fill = 'petal_width'),
                alpha=0.5,
                size = 2,
                stroke = 0.2) 
    + geom_smooth(
         aes(x='petal_width', y = 'petal_length')
    )
    + labs(title='Anchura y altura del ptalo', 
          x = 'Anchura Ptalo',
          y = 'Altura Ptalo')
    + facet_wrap('~petal_width', ncol = 4)
)
```

### Mostrar valores mximos, mnimos cuartiles y la mediana

```python
(
     ggplot(df) 
    + geom_boxplot(
         aes(x='petal_width', y = 'petal_length', fill = 'type'),  
         alpha=0.4, 
         outlier_alpha = 1, 
         outlier_color ='red'
     )
    + labs(title='Anchura y altura del ptalo', 
          x = 'Anchura Ptalo',
          y = 'Altura Spalo',
          color = 'Ancho Ptalo')
)
```
