## Superponer diferentes variables

```python
import numpy as np

valores = np.linspace(0,2,100)

plt.plot(valores)
plt.plot(valores**2)
plt.plot(valores**3)

plt.xlabel('Item')
plt.ylabel('Valor')

plt.title('Transformacin de variables')
plt.show()
```

```python
plt.plot(valores, color='red', label='Lineal')
plt.plot(valores**2, color = 'blue', label = 'Cuadrtica')
plt.plot(valores**3, color = 'green', label = 'Cbica')

plt.xlabel('tem')
plt.ylabel('Valor')
plt.title("Transformacin de una varaible")
plt.legend()

plt.show()
```

```python
plt.plot(valores, color = 'red', label = 'Lineal', marker='.') 
plt.plot(valores**2, color = 'blue', label = 'Cuadrtica', marker = '^' )
plt.plot(valores**3, color = 'green', label = 'Cbica', marker='s')
 
plt.xlabel('tem')
plt.ylabel('Valor')
plt.title("Transformacin de una varaible")

plt.legend()
plt.show()
```

```python
t = np.arange(0., 5., 0.2)

plt.plot(t, t, '--')
plt.plot(t, t**2, 's')
plt.plot(t, t**3, '^')

plt.show()
```

```python
def balance(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

my_array_one = np.arange(0.0, 5.0, 0.1)
my_array_two = np.arange(0.0, 5.0, 0.02)

plt.subplot(211)
plt.plot(my_array_one, balance(my_array_one), 'bo')
plt.plot(my_array_two, balance(my_array_two), 'k')

plt.subplot(212)
plt.plot(my_array_two, np.cos(2*np.pi*my_array_two), 'r--')
plt.show()
```

### Histograma

```python
mu = 150 # Media de la distribucin
sigma = 15 #Desviacin estandar de la distribucin
x = mu + sigma * np.random.randn(1000) #Distribucin nomal de 1000 elementos

num_bins = 100

# Mostramos valores
n, bins, patches = plt.hist(x, num_bins, density=1, color='lightgreen')

# Distribucin normal
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

plt.plot(bins, y, 'b--')
plt.xlabel('Frecuencia')
plt.ylabel('Valor')

plt.show()
```

## Diagramas

```python
# Diagrama de barras 1
import pandas as pd
fechas = pd.date_range('20200101', periods=12)

air_quality = pd.DataFrame(np.random.randint(10,50,24).reshape(12,2), 
                            index=fechas, columns=['AQI Madrid', 'AQI Barcelona'])
air_quality
```

```python
# Diagrama de barras 2
air_quality.plot(kind="barh", title="Air Quality Index")
```

```python
# Diagrama de barras 3
air_quality.plot(kind="barh", subplots=True)
```

```python
# Diagrama de barras, ejemplo con listas.

# Datos sobre temperatura
temp_max = [35, 38, 29, 27]
temp_min = [23, 22, 21, 18]
n_groups = len(temp_max)

index = np.arange(n_groups)
max = plt.bar(index, temp_max, color='tomato', label='T.Max')
min = plt.bar(index, temp_min, color='lightgreen', label='T.min')

plt.xlabel('Ciudades')
plt.ylabel('Temperaturas')
plt.title('Temperatura por ciudad')

plt.xticks(index , ('Madrid', 'Barcelona', 'Asturias', 'Burgos'))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Diagrama Circular
politics = ('PSOE', 'PP', 'VOX', 'PODEMOS', 'ERC', 'Cs')
votos = [35, 28, 17, 12, 5, 3 ]

explode = (0.2, 0.0, 0, 0, 0, 0)  # Resaltar hacia afuera una porcin.

plt.pie(votos, explode=explode, labels=politics, autopct='%1.1f%%',
         shadow=True, startangle=90, colors = ('r', 'b', 'g', 'purple', 'yellow', 'orange' ))

plt.axis('equal') # Centrar el grfico
plt.show()
```
